# Terms
* Interpretability - Understanding the model's behavior by a human
* Mechanistic Interpretability - Aims to reverse-engineer model computations into human-understandable components.
* Indirect Object Identification (IOI) - Identifying indirect objects in a sentence. 
    * NLP task
    * John gave Mary a book," "Mary" is the indirect object
* In the Wild - Examining how the GPT-2 Small model performs the task of Indirect Object Identification (IOI) under conditions that are more typical of how the model would be used in real-world applications.
* Circuit - Attention heads within the GPT-2 model that are crucial for performing the task of identifying indirect objects in sentences.

# Methodology
* Replicate GPT-2 Small: 12-layer, 768-hidden, 12-heads, 117M parameters
* 
* Specific task focus: Indirect Object Identification
* Path Patching - is tracing logits back through the network to identify which neurons, layers, or attention heads are most responsible for that logits value.

## Side Note
* Garçon models expose a set of “probe points”

# A Circuit implements this algorithm:
1. Identify all previous names in the sentence (Mary, John, John).
2. Remove all names that are duplicated (in the example above: John). 
3. Output the remaining name.



# Validation Metrics
Three criteria are used to validate the identified circuit: faithfulness, completeness, and minimality.
